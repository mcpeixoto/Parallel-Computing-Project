```python
import os
from os.path import join, basename
import matplotlib.pyplot as plt
import pandas as pd
from glob import glob
from tqdm.notebook import tqdm
```

```python
# For all data in data folder
for file_path in tqdm(glob(join('data', '*.txt'))):
    print("-"*5, f"Plotting {basename(file_path)}", "-"*5)
    
    # Get file name
    file_name = basename(file_path).split('.')[0]

    # Load data
    data = pd.read_csv(file_path, sep='\t')
    
    # Get columns
    columns = data.columns

    # Separate columns
    x_axis = columns[0]
    columns_to_plot = columns[1:]

    # Plot and save figure
    for column in columns_to_plot:
        plt.plot(data[x_axis], data[column], label=column)

        # Labels
        plt.xlabel(x_axis)
        plt.ylabel(column)

        # Scales
        plt.yscale('log')
        plt.xscale('log')

        # Display
        plt.legend()
        
        # Save image
        plt.savefig(join(os.getcwd(), 'graficos', f'{file_name}_{x_axis}_{column}.png'))
        plt.show()

```
